```python


# Clean column names
df.columns = df.columns.str.strip()

# Ensure 'Trade amount' is numeric (it seems it already is)
df['Trade amount'] = pd.to_numeric(df['Trade amount'], errors='coerce')

# Group by 'Security Name' and calculate Buy and Sell amounts
grouped = df.groupby(['Security Name', 'Buy / Sell'])['Trade amount'].sum().unstack(fill_value=0)
grouped.columns.name = None
grouped = grouped.rename(columns={'BUY': 'Buy Amount', 'SELL': 'Sell Amount'})

# Calculate Total Trade
grouped['Total Trade'] = grouped['Buy Amount'] + grouped['Sell Amount']

# Sort by Total Trade descending
grouped = grouped.sort_values('Total Trade', ascending=False)

# Display the results
print(grouped)

# Export to CSV
grouped.to_csv('trade_summary.csv')

# Visualization
plt.figure(figsize=(12, 6))
grouped[['Buy Amount', 'Sell Amount']].plot(kind='bar', stacked=True)
plt.title('Buy and Sell Amounts by Security')
plt.xlabel('Security Name')
plt.ylabel('Amount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('trade_summary_plot.png')
plt.close()

# Top 5 securities by total trade
top_5 = grouped.nlargest(5, 'Total Trade')
print("\nTop 5 Securities by Total Trade:")
print(top_5)

# Calculate and display overall statistics
total_buy = grouped['Buy Amount'].sum()
total_sell = grouped['Sell Amount'].sum()
total_trade = grouped['Total Trade'].sum()

print(f"\nOverall Statistics:")
print(f"Total Buy Amount: {total_buy:,.2f}")
print(f"Total Sell Amount: {total_sell:,.2f}")
print(f"Total Trade: {total_trade:,.2f}")
print(f"Net Position (Buy - Sell): {total_buy - total_sell:,.2f}")
```

```python
sorted_data = final_stockdf.sort_values('Net Ratio', ascending=False).head(10)
norm = plt.Normalize(sorted_data['Net Ratio'].min(), sorted_data['Net Ratio'].max())

# Generate colors from dark to light green
colors = plt.cm.Greens(norm(sorted_data['Net Ratio']))
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    x='Security Name',
    y='Net Ratio',
    data=sorted_data,
    palette=colors
)
plt.title('Top 10 Securities by Net Ratio')
plt.xlabel('Security Name')
plt.ylabel('Net Ratio')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```
